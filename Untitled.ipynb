```python
import pickle
import os
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# with only 2D memview fixed
with open('./pr2d_results.pickle', 'rb') as handle:
    pr_results = pickle.load(handle)
# with open('./pr_results2.pickle', 'rb') as handle:
#     pr_results_2 = pickle.load(handle)
```

```python
# with 2D memview and 1D memview fixed
with open('./pr2d1d_results.pickle', 'rb') as handle:
    pr2_results = pickle.load(handle)
# with open('./pr2_results2.pickle', 'rb') as handle:
    # pr2_results_2 = pickle.load(handle)
```

```python
with open('./prmain_results.pickle', 'rb') as handle:
    main_results = pickle.load(handle)
# with open('./main_results2.pickle', 'rb') as handle:
#     main_results_2 = pickle.load(handle)
```

```python
print(pr_results.keys())
# print(pr_results_2.keys())
```

```python
key = 'kmeans_speed'
```

```python
sns.set_context('talk')
fig, ax = plt.subplots()

for branch, results in zip(['pr', 'pr2', 'main'], [pr_results, pr2_results, main_results]):
    speed = results[key]
    
    ax.plot(speed, marker='.', label=branch)
    
ax.legend()
ax.set(
    xlabel='Index',
    ylabel='Time',
    title='KMeans Runtime Over \nIncreasing Density',
)
```

```python
fig, ax = plt.subplots()
ax.plot(np.array(pr2_results[key]) - np.array(main_results[key]), marker='.', label='2D&1D-Main')
ax.plot(np.array(pr_results[key]) - np.array(main_results[key]), marker='.', label='2D-Main')
ax.plot(np.array(pr2_results[key]) - np.array(pr_results[key]), marker='.', label='2D&1D-2D')
ax.axhline([0], ls='--', label='No change')
ax.legend()
ax.set(
    xlabel='Index of Run',
    ylabel='Delta Time',
    title='Delta Runtime Over \nIncreasing Sample and Feature Size',
)
```

```python
samples_range = np.linspace(50, 5000, 10).astype(int)
features_range = np.linspace(150, 50000, 5).astype(int)

sample_arr = []
feature_arr = []
for samp in samples_range:
    for feat in features_range:
        sample_arr.append(samp)
        feature_arr.append(feat)
        
sample_arr = np.array(sample_arr).reshape(10, 5)
feature_arr = np.array(feature_arr).reshape(10, 5)
print(samples_range)
print(features_range)
```
